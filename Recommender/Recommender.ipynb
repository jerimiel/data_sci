```python
#Tie the recommender all up in a function
def recommend(id):
    try:
        user=table.loc[id].dropna()
        books=list(user.index)
        user9=corr[books].dropna()
        for i in user9.columns:
            user9[i]=(user9[i]*user[i])/2
        ans=user9.sum(axis=1)/len(user9.columns)
        ans=ans.sort_values(ascending=False)
        ans=((ans-min(ans))/(max(ans)-min(ans))).head(5)
    except:
        ans=pd.Series(None,dtype=float)
    return ans
```

```python
#tieing up
def recs(id):
    try:
        similar_users=corr0[id].dropna().sort_values(ascending=False).head(5)
        cols=list(similar_users.index)
        recs=(pivot[cols].sum(axis=1)/len(cols)).sort_values(ascending=False)
        recs= ((recs-min(recs))/(max(recs)-min(recs))).head()
    except:
        recs=pd.Series(None,dtype=float)
    return recs
```

```python
recs(1)
```

```python
#putting them all together
def final(id):
    users=recs(id)
    ratings=recommend(id)
    users=pd.DataFrame(users)
    ratings=pd.DataFrame(ratings)
    tot=pd.concat([ratings,users],axis=0)
    return tot[0].sort_values(ascending=False)
```
